```python
#Çalışanları (EmployeeID) isimleri ile birlikte yaptıkları toplam satışa göre sıralayınız.
```

```python
%%sql
select Employees.EmployeeID, count(Orders.OrderID)
from Employees 
left join Orders on Employees.EmployeeID = Orders.EmployeeID 
group by Employees.EmployeeID
```

```python
%%sql
SELECT em.FirstName, SUM(p.Price) toplam_satis 
FROM Orders o 
LEFT JOIN OrderDetails od on o.OrderID = od.OrderID 
LEFT JOIN Products p on od.ProductID = p.ProductID 
LEFT JOIN Employees em on o.EmployeeID = em.EmployeeID 
GROUP BY o.EmployeeID 
ORDER BY toplam_satis DESC;
```

```python
#Germany'den verilen siparişlerin kategorilerine göre ortalama fiyatı.
```

```python
%%sql
select c.CategoryName, AVG(p.Price)  
from Orders o 
left join OrderDetails od on od.OrderID = o.OrderID
left join Products p on od.ProductID = p.ProductID
left join Categories c on p.CategoryID = c.CategoryID
left join Customers cs on o.CustomerID = cs.CustomerID
where cs.Country = 'Germany' 
group by CategoryName

```

```python
#Almanya ya da USA'den verilen siparişlerin kategorilerine göre ortalama fiyatları nelerdir?
```

```python
%%sql
SELECT ca.CategoryName, AVG(p.Price) 
FROM Orders o 
LEFT JOIN OrderDetails od on o.OrderID = od.OrderID 
LEFT JOIN Products p on od.ProductID = p.ProductID 
LEFT JOIN Categories ca on p.CategoryID = ca.CategoryID 
LEFT JOIN Customers cu on o.CustomerID = cu.CustomerID 
WHERE cu.Country = 'USA' or cu.Country = 'Germany' GROUP BY CategoryName;
```

```python
#Haziran Temmuz Ağustos aylarında verilen siparişlerin ortalama fiyatı nedir?
```

```python
%%sql
select avg(Price) 
from Orders o 
left join OrderDetails od on o.OrderID = od.OrderID 
left join Products p on od.ProductID = p.ProductID
where OrderDate like '%07%' or OrderDate like '%08%' or OrderDate like '%09%';
```

```python
#Müşterilerin 1997 yılına ait siparişlerinin maksimum miktarlarını bulup müşterilerin isimleri ile birlikte büyükten küçüğe sıralayınız.
```

```python
%%sql
select CustomerName, MAX(Price) maks 
from Orders o
left join Customers c on o.CustomerID = c.CustomerID
left join OrderDetails od on o.OrderID = od.OrderID
left join Products p on od.ProductID = p.ProductID
where OrderDate like '%1997%'
GROUP BY CustomerName ORDER BY maks DESC
```

```python
#Sipariş yılı 1997 olan siparişleri alan çalışanları sipariş alma sayılarına göre sıralayınız.
```

```python
%%sql
select count(o.OrderID), e.EmployeeID, e.FirstName
from Orders o 
left join Employees e on o.EmployeeID = e.EmployeeID
where OrderDate like '%1997%' 
group by e.EmployeeID
order by count(o.OrderID)
```
